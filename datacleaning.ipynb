```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\dhrit\\Downloads\\NFL Play by Play 2009-2016 (v3).csv", low_memory=False)
print( df.shape)
display(df.head())
```

```python
df = df.drop_duplicates()
print( df.shape)
```

```python
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip().str.upper()
display(df.head(5))
```

```python
placeholders = ["NA", "N/A", "NONE", "NULL", "NAN", "?", "-", "--", "UNKNOWN", "ERROR"]
for col in df.columns:
    df[col] = df[col].replace(placeholders, np.nan)
missing_table = df.isna().sum().sort_values(ascending=False).head(10).reset_index()
missing_table.columns = ["Column", "Missing Values"]
display(missing_table)
```

```python
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
dtypes_table = df.dtypes.head(10).reset_index()
dtypes_table.columns = ["Column", "Data Type"]
display(dtypes_table)
```
